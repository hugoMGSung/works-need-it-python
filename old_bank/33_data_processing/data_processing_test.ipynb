```python
df3 = pd.DataFrame(np.random.randint(50, 100, size=(5,5)),
                   index=[['3학년' for i in range(5)],
                          [str(i)+'반' for i in range(1, 6)]],
                   columns=['국어','영어','수학','사회','과학'])
print(df3)
```
