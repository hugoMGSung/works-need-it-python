# Matplotlib 다시

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
flg = plt.figure()

plt.plot(x, np.sin(x), '-') # 실선으로 sin 그래프 
plt.plot(x, np.cos(x), '--') # 점선으로 cos 그래프
# plt.plot(x, np.tan(x), '--^') # 점선과 삼각형으로 tan 그래프
```

```python
# 이미지로 저장
flg.savefig('sin_cos.png')
```
