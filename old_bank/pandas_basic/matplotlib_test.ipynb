```python
import matplotlib.pyplot as plt
```

```python
# 라인 플롯
plt.title('line plot')
plt.plot([1,2,3,4,5],[2,4,8,6,9])
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
# 라인 추가
plt.title('line plot 2')
plt.plot([1,2,3,4,5],[2,4,8,6,9]) 
# 두번째 라인 
plt.plot([1.5, 2.5, 3.5, 4.5], [11, 24, 30, 11])
plt.xlabel('hours')
plt.ylabel('score')
plt.legend(['A group', 'B group'])
plt.show()
```
