#### Bark: Suno의 Transformer 기반 Text-to-Audio 모델
- Suno가 만든 트랜스포머 기반의 텍스트-오디오 변환 모델
- 음악, 배경 소음, 간단한 음향 효과 등 기타 오디오뿐만 아니라 매우 사실적인 다국어 음성을 생성

```python
# 설치
!pip install git+https://github.com/suno-ai/bark.git
```

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio
```

```python
# download and load all models
preload_models()
```

- GPU가 사용되지 않으며 긴시간 소요

```python
# generate audio from text
text_prompt = """
     파이썬 능력을 한 단계 더 끌어올리고 싶으신가요? 그렇다면, 파이썬 생태계의 넓은 세상을 경험해보세요.
"""
audio_array = generate_audio(text_prompt)
```

```python
# save audio to disk
write_wav("./output/bark_generation.wav", SAMPLE_RATE, audio_array)
```

```python
# play text in notebook
Audio(audio_array, rate=SAMPLE_RATE)
```

#### KoNLPy: 파이썬 한국어 NLP
- https://konlpy.org/ko/latest/
- 한국어 처리를 위한 파이썬 라이브러리

##### 설치

- JAVA_PATH가 설치 된 JRE bin 폴더를 지정

```python
# 설치
!pip install konlpy
```

```python
from konlpy.tag import Okt
```

```python
# Okt 형태소 분석기 인스턴스 생성
okt = Okt()

# 텍스트 정의
text = "한국어 분석을 시작합니다. 재미있어요!"
```

```python
# 형태소 분석 (토큰화)
morphs = okt.morphs(text)
print("형태소:", morphs)
```

```python
# 품사 태깅
pos = okt.pos(text)
print("품사 태깅:", pos)
```

```python
# 명사 추출
nouns = okt.nouns(text)
print("명사:", nouns)
```

#### SpeechRecognition: 음성 인식 라이브러리
- https://pypi.org/project/SpeechRecognition/
- 음성 인식 라이브러리. 여러 음성 인식 서비스를 지원하며, 사용자의 음성을 텍스트로 변환하는 데 사용

```python
# 설치
!pip install SpeechRecognition
```

```python
# 예제코드(실행하지 않음)
import speech_recognition as sr

# 음성 인식 객체 생성
recognizer = sr.Recognizer()

# 마이크에서 입력받기
with sr.Microphone() as source:
    print("말하세요...")
    audio = recognizer.listen(source)

# Google Web Speech API를 사용하여 음성 인식
try:
    print("인식된 텍스트: " + recognizer.recognize_google(audio, language='ko-KR'))
except sr.UnknownValueError:
    print("Google Web Speech API가 당신의 말을 이해하지 못했습니다.")
except sr.RequestError as e:
    print(f"Google Web Speech API 서비스에 문제가 발생했습니다; {e}")
```
