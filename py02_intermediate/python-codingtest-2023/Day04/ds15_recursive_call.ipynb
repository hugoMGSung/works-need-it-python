## 재귀호출 (Recursive Call)

```python
def printAll(n):
    print(f'n = {n}')
    return n

def plus(x, y):
    answer = x + y
    result = printAll(answer)
    return result

val = plus(3, 4)
print(val)
```

```python
count = 3

def openBox():
    global count
    print('종이상자를 엽니다. ^^')
    count -= 1
    if count == 0:
        print('반지를 넣고 반환합니다*****')
        return
    
    openBox() # 자기자신을 다시 호출
    print('종이 상자를 닫습니다.')

openBox()
```

### 재귀호출 하지않고 반복문으로 계산

```python
sumValue = 0

for n in range(10, 0, -1):
    sumValue += n

print('10+9+8+...+1 = ', sumValue)
```

### 재귀호출로 덧셈

```python
def addNumber(num):
    if num <= 1:
        return 1
    else:
        return num + addNumber(num - 1)

print(addNumber(100))
```

### 팩토리얼 반복문

```python
factValue = 1
for n in range(1300, 0, -1):
    factValue *= n

print('10*9*8..*2*1', factValue)
```

### 팩토리얼 재귀호출

```python
def factorial(num):
    if num <= 1:
        return 1
    else:
        return num * factorial(num - 1)
    
print('10! = ', factorial(1200))
```

### 우주선 발사

```python
import time
def countdown(n):
    if n == 0:
        print('우주선 발사!!')
    else:
        print(n)
        time.sleep(1)
        countdown(n - 1)

countdown(10)
```

```python
def printStar(n):
    if n > 0:
        printStar(n-1)
        print('★' * n)
        

printStar(5)
```

### 구구단 재귀호출

```python
def gugudan(dan, num):
    print(f'{dan} x {num} = {dan*num:2d}')
    if num < 9:
        gugudan(dan, num+1)
    
for dan in range(2, 10):
    print(f'{dan}단 시작 ------')
    gugudan(dan, 1)
```

#### N제곱

```python
tab = ''
def pow(x, n):
    global tab
    tab += ' '
    if n == 0:
        return 1
    print(tab, f'{x}x{x}^{n}-{1}')
    return x * pow(x, n-1)

print('2^10')
print('답 --> ', pow(2, 10))
```

### 피보나치 수열

```python
def fibonacci(n):
    if n == 0 or n == 1 or n == 2: return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
print('피보나치 수 --> 0 1', end=' ')
for i in range(2, 39):
    print(fibonacci(i), end=' ')
```

### 회문 여부

```python
def palindrome(pStr): # kayak
    if len(pStr) <= 1:
        return True
    
    if pStr[0] != pStr[-1]: # k == k a == a
        return False
    else:       
        return palindrome(pStr[1:len(pStr)-1])
```

```python
strAry = ['reaver', 'kayak', 'Borrow or rob', '주유소의 소유주', 
          '야 너 이번 주 주번이 너야', '살금 살금 살']
```

```python
for testStr in strAry:
    print(testStr, end='--> ')
    testStr = testStr.lower().replace(' ', '') # 영어대문자X, 공백X
    if palindrome(testStr) == True:
        print('회문입니다!')
    else:
        print('회문아니오 ㅡㅡ;')
```
